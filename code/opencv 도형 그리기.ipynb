```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

cv2.line(image, start, end, color, thickness) : 하나의 직선을 그리는 함수
- start : 시작 좌표(2차원)
- end : 종료 좌표(2차원)
- thickness : 선의 두께

```python
image_path = 'C:/Users/sungyun/Desktop/etc/img/sibadog.jpg'
# image = np.full((512,512,3), 255, np.uint8)
image = cv2.imread(image_path, cv2.IMREAD_COLOR)
image = cv2.line(image, (0, 100), (200, 200), (0, 0, 255), 3)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

cv2.rectangle(image, start, end, color, thickness) : 하나의 사각형을 그리는 함수

- start : 시작 좌표
- end : 종료 좌표
- thickness : 선의 두께 (-1 은 안을 채운다)

```python
image = cv2.imread(image_path, cv2.IMREAD_COLOR)
image = cv2.rectangle(image, (20,20), (100,100), (0, 0, 255), -1) # 마이너스는 사각형을 가득 채운다

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

cv2.circle(image, center, radian, color, thickness) 하나의 원을 그리는 함수

- center : 원의 중심
- radian : 반지름
- thickness : 선의 두께 (-1 은 안을 채운다)

```python
image = cv2.imread(image_path, cv2.IMREAD_COLOR)
image = cv2.circle(image, (50,50), 30, (255, 0, 0), -1)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

cv2.polylines(image, points, is_closed, color, thickness) : 하나의 다각형을 그리는 함수

- points : 꼭지점들
- is_closed : 닫힌 도형 여부
- thickness : 선의 두께 (-1 은 채우기)

```python
image = cv2.imread(image_path, cv2.IMREAD_COLOR)
points = np.array([[125,75], [200, 75], [200, 180], [125,180]])
image = cv2.polylines(image, [points], True, (0, 0, 255), 4)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

cv2.putText(image, text, position, font_type, font_scale, color) : 하나의 텍스트를 그리는 함수

- position : 텍스트가 출력될 위치
- font_type : 글씨체
- font_scale : 글씨 크기 가중치

```python
image = cv2.imread(image_path, cv2.IMREAD_COLOR)
image = cv2.putText(image,'hi siba', (125,50), cv2.FONT_HERSHEY_DUPLEX, 1, (0,0,0))

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
