# opencv 픽셀 다루기

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
image_path = 'C:/Users/sungyun/Desktop/etc/img/sibadog.jpg'
image = cv2.imread(image_path, cv2.IMREAD_COLOR)
image2 = cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)
print(image.shape)
print(image.size)
print(image2.shape)
print(image2.size)
# 이미지의 크기 및 픽셀 확인

px = image[100,100] 
# 불러온 이미지의 [100 , 100] 의 픽셀을 가져올 수 있다. B G R 순으로 불러온다.
# gray scale 일 경우엔 B G R 순이 적용되지 않는다.
print(px)
```

```python
px2 = np.concatenate((image2,image2), axis=0)
px3 = np.concatenate((px2,px2), axis=1) 
```

```python
plt.imshow(cv2.cvtColor(px3, cv2.COLOR_BGR2RGB))
plt.show()
# 시바견을 합쳐보았다. 배열을 그대로 유지해 준다면 원하는 대로 위치 조정도 가능할 거 같다.
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
##### start_time = time.time()
for i in range(50, 100):
    for j in range(10, 100):
        image[i, j] = [0,0,0]
        
print('=== %s seconds ===' % (time.time() - start_time))

start_time = time.time()
image[0:100, 0:100] = [0,0,0]
print('=== {} seconds ==='.format(time.time()-start_time))
# for 문으로 바꿀 순 있지만 인덱싱이 빠른 속도로 할 수 있다.

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

# 픽셓들의 값들을 바꾸어 주면 이미지의 부분이 바뀌게 된다.
```

```python
image = cv2.imread(image_path, cv2.IMREAD_COLOR)

roi = image[80:180, 110:210]

image[0:100, 0:100] = roi

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

# 원하는 이미지를 따로 잘라서 가져올 수 있다.
```
