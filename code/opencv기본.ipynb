# opencv2 기본

```python
import cv2
import matplotlib.pyplot as plt
```

- jupyter나 colab은 matplotlib을 이용해 출력 할 수 있다.
- pycharm 같은 개발환경에선 cv2.imshow()를 이용해 windows 창을 띄어 이미지를 출력 할 수 있다.

```python
img_basic = cv2.imread('C:/Users/sungyun/Desktop/etc/img/sibadog.jpg', cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_BGR2RGB))
plt.show()

img_gray = cv2.imread('C:/Users/sungyun/Desktop/etc/img/sibadog.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_BAYER_BG2BGR))
plt.show()
```

cv2.imread(imgfile_name, flag)
* imgfile_name : 불러올 이미지 파일
* flag : 이미지를 읽는 방법 설정

1. IMREAD_COLOR : 이미지를 컬러로 불러온다, 투명한 부분은 무시한다.
2. IMREAD_GRAYSCALE : 이미지를 Grayscale로 읽기, 검은색과 하얀색으로만 읽는다, 즉 그냥 흑백
3. IMREAD_UNCHANGED : 이미지를 컬러로 불러온다, 투명한 부분도 읽어온다.

반환 값 : Numpy 객체 (행, 열, 색상 : 기본 BGR)

```python
cv2.imshow('title', img_basic)
cv2.waitKey(0)  # 키보드 입력을 처리하는 함수, 키보드 입력으로 출력된 창을 종료할 수 있다. 0은 무한대기를 의미한다.
```
